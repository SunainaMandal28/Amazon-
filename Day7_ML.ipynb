```python
from FlagEmbedding import FlagReranker
reranker = FlagReranker('BAAI/bge-reranker-v2-m3', use_fp16=True) # Setting use_fp16 to True speeds up computation with a slight performance degradation
```

```python
scores = reranker.compute_score([['what is panda?', 'hi'], ['what is panda?', 'The giant panda (Ailuropoda melanoleuca), sometimes called a panda bear or simply panda, is a bear species endemic to China.']])
print(scores) # [-8.1875, 5.26171875] #re means relevances
```

#Pipeline Function
# It connects a model with its necessary preprocessing and postprocessing steps, allowing us to drecting input any text and get an intelligible answer

```python
from transformers import pipeline
```

```python
#Text Pipelines
```

```python
classifier = pipeline("sentiment-analysis")
classifier(" I am Happy")
```

```python
classifier = pipeline("sentiment-analysis")
classifier(" I am traumatized")
```

```python
classifier = pipeline("sentiment-analysis")
classifier(" I am traumatized but happy")
```

```python
classifier = pipeline("sentiment-analysis")
classifier(" I am happy but traumatized")
```

```python
classifier = pipeline("sentiment-analysis")
classifier([" I am happy" ," I am traumatized"])
```

Zero-shot-classification

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

```python
#Example 2
classifier(
    "The government announced a new policy for public transport",
    candidate_labels=["politics", "technology", "health"],
)
```

Text Generation

```python
generator = pipeline("text-generation")
generator("In this course, we will teach you how to")
```

Text generation using distilgpt2 Model

```python
generator = pipeline("text-generation", model="distilgpt2") # yesma chaii customize garnu milxa like max length and all
generator (
    "In this course, we will teach you how to",
    max_length=30,
    num_return_sequences=2,
)
```

Text completion (mask filling)

```python
unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=2)
```

ner (Name - Entity - Recognition) -> Token Classification

```python
ner = pipeline("ner", grouped_entities=True)
ner("My name is Sunaina and i work at HuggingFace in Bangalore")
```
